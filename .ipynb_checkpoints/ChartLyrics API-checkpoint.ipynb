```python
song_df = pd.DataFrame(columns=['lyric_id', 'lyric_check_sum', 'artist', 'song'])

for item in data.iterfind('*'):
    artist = item.findtext('{http://api.chartlyrics.com/}Artist')
    song = item.findtext('{http://api.chartlyrics.com/}Song')
    lyric_id = item.findtext('{http://api.chartlyrics.com/}LyricId')
    lyric_cs = item.findtext('{http://api.chartlyrics.com/}LyricChecksum')

    song_df = song_df.append({'lyric_id': lyric_id, 
                              'lyric_check_sum': lyric_cs, 
                              'artist': artist, 
                              'song': song}, 
                              ignore_index=True)
    
song_df
```

## Get lyric text

```python
selected_lyric_id = song_df.loc[0, 'lyric_id']
selected_lyric_id
```

```python
selected_lyric_cs = song_df.loc[0, 'lyric_check_sum']
selected_lyric_cs
```

```python
base_lyric_url = 'http://api.chartlyrics.com/apiv1.asmx/GetLyric?'
qs_args = {
    'lyricId': selected_lyric_id,
    'lyricCheckSum': selected_lyric_cs
}
selected_url = base_lyric_url + urllib.parse.urlencode(qs_args)
selected_url
```

```python
selected_response = urllib.request.urlopen(selected_url)
selected_content = selected_response.read().decode('ascii')
selected = ET.fromstring(selected_content)
```

```python
selected_lyrics = selected.find('{http://api.chartlyrics.com/}Lyric').text
selected_lyrics
```

```python
selected_lyrics = selected_lyrics.replace('\n', ' ')
selected_lyrics
```

## Apply model

```python
def tokenize(text):
    tokens = word_tokenize(text)
    lemmatizer = WordNetLemmatizer()
    
    clean_tokens = []
    for tok in tokens:
        clean_tok = lemmatizer.lemmatize(tok).lower().strip()
        clean_tokens.append(clean_tok)
        
    return clean_tokens
```

```python
model = joblib.load('data/classifier.pkl')
```

```python
classification_labels = model.predict([selected_lyrics])[0]
classification_labels
```
